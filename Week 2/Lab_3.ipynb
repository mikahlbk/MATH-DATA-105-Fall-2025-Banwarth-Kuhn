```python
#define a function that will do the Rate of Change computation for us
def compute_RoC_func(final_input_val,start_input_val,final_output_val,start_output_val):
    input_change = final_input_val-start_input_val#denominator, i.e. change in input
    output_change = final_output_val-start_output_val#numerator, i.e. change in ouput
    return output_change/input_change#change in output/change in input
```
