```python
import numpy as np
import matplotlib.pyplot as plt

# Given data
dollars_spent_social_media = np.array([18.7270, 47.5357, 36.5997, 29.9329, 7.8009])
dollars_spent_newspaper = np.array([7.7997, 2.9042, 43.3088, 30.0558, 35.4036])
sales_revenue = np.array([6.1964, 13.7375, 11.5585, 9.5579, 6.8143])

#############################################################################################
# SET PARAMETERS
#############################################################################################
m1 = .1
m2 = .005
b = .5

#############################################################################################
# Plot model vs observed data
#############################################################################################
f1 = lambda M,N:m1*M+m2*N+b
# Create first plot: Sales vs. Social Media Spend
plt.figure()
plt.scatter(dollars_spent_social_media, sales_revenue, marker = 'o',color='blue',label = "Observed Data")
plt.scatter(dollars_spent_social_media, f1(dollars_spent_social_media,dollars_spent_newspaper,), color='red',label = "Predicted")
plt.xlabel("Dollars Spent on Social Media")
plt.ylabel("Sales Revenue")
plt.title("Sales vs. Social Media Spending")
plt.grid(True)
plt.legend()
plt.tight_layout()

#############################################################################################
# Plot model vs observed data
#############################################################################################
# Create second plot: Sales vs. Newspaper Spend
plt.figure()
plt.scatter(dollars_spent_newspaper, sales_revenue, marker = 'o',color='blue',label = "Observed Data")
plt.scatter(dollars_spent_newspaper, f1(dollars_spent_social_media,dollars_spent_newspaper,), color='red',label = "Predicted")
plt.xlabel("Dollars Spent on Newspaper")
plt.ylabel("Sales Revenue")
plt.title("Sales vs. Newspaper Spend")
plt.grid(True)
plt.legend()
plt.tight_layout()

plt.show()
```
