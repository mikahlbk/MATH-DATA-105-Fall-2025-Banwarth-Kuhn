$\text{\textbf{Derivative Rules:}}$

Calculating the derivative of a function using the definition is a tedious process involving a lot of algebra. In practice, when we want to find the derivative function, we use well-known rules. Each of these rules can be derived using the definition. If you are curious a quick internet search or ChatGpt convo can give you a good overview of how to derive each rule.

We will go over some of the most important ones now.

$\text{\textbf{The Derivative of a Constant Function:}}$

Suppose $f(t) = c$ for any number $c$, then
$$f'(t) = 0.$$

We can also write this using Leibniz notation,
$$\frac{df}{dt}(t) = 0.$$

Or the defintion of the derivative
$$\lim_{\Delta t \to 0} \frac{f(t+\Delta t) -f(t)}{\Delta t} = 0.$$


```python
#Example
import sympy as sp

# Define the input variable
t = sp.Symbol('t')

# Define the function
f = 5
print("Function:", f)
# Turn the function into a Python function
f_func = sp.lambdify(t, f, 'numpy')

# Take the derivative
f_prime = sp.diff(f, t)
print("Derivative Function:", f_prime)
# Turn the derivative into a Python function
f_prime_func = sp.lambdify(t, f_prime, 'numpy')

#Plot it
import matplotlib.pyplot as plt
import numpy as np
t_vals = np.linspace(-20, 20, 100)
plt.plot(t_vals, np.full_like(t_vals, f_func(t_vals)), color="black", label="f(t)")
plt.plot(t_vals, np.full_like(t_vals, f_prime_func(t_vals)), color="blue", label="f'(t)")
plt.legend()
plt.show()
```

If $f(t)$ and $g(t)$ are two functions, then $$\frac{d}{dt}(f + g) = \frac{d}{dt}(f)+\frac{d}{dt}(g)$$.

Example: Suppose $f(t) = 2t + 10$. Then $$\frac{df}{dt}(t) = \frac{d}{dt}(2t) + \frac{d}{dt}(10) = 2 + 0 = 2.$$

Do $f_1(t) = 2t+5$ and $f_2(t) = 2t-5$ have the same derivative?

```python
#Example
import sympy as sp

# Define the input variable
t = sp.Symbol('t')

# Define the function
f1 = 2*t+5
print("Function:", f1)
# Turn the function into a Python function
f1_func = sp.lambdify(t, f1, 'numpy')

# Define the function
f2 = 2*t-5
print("Function:", f2)
# Turn the function into a Python function
f2_func = sp.lambdify(t, f2, 'numpy')

# Take the derivative
f1_prime = sp.diff(f1, t)
print("Derivative Function:", f1_prime)
# Turn the derivative into a Python function
f1_prime_func = sp.lambdify(t, f1_prime, 'numpy')

# Take the derivative
f2_prime = sp.diff(f2, t)
print("Derivative Function:", f2_prime)
# Turn the derivative into a Python function
f2_prime_func = sp.lambdify(t, f2_prime, 'numpy')

#Plot it
import matplotlib.pyplot as plt
import numpy as np
t_vals = np.linspace(-5, 5, 100)
plt.plot(t_vals, np.full_like(t_vals, f1_func(t_vals)), color="black", label="f1(t)")
plt.plot(t_vals, np.full_like(t_vals, f1_prime_func(t_vals)), color="blue", label="f1'(t)")
plt.plot(t_vals, np.full_like(t_vals, f2_func(t_vals)), color="red",linestyle = "--", label="f2(t)")
plt.plot(t_vals, np.full_like(t_vals, f2_prime_func(t_vals)), color="green",linestyle = "--",label="f2'(t)")
plt.legend()
plt.show()
```

$\text{\textbf{The Derivative of Exponential Functions:}}$

Suppose $f(t) = e^t$, then
$$f'(t) = e^t = 1*f(t) = k*f(t).$$

We can also write this using Leibniz notation,
$$\frac{df}{dt}(t) =e^t.$$

Or the defintion of the derivative
$$\lim_{\Delta t \to 0} \frac{f(t+\Delta t) -f(t)}{\Delta t} = e^t.$$

```python
#Example
import sympy as sp
import numpy as np

# Define the input variable
t = sp.Symbol('t')

# Define the function
f = sp.exp(t)
print("Function:", f)
# Turn the function into a Python function
f_func = sp.lambdify(t, f, 'numpy')

# Take the derivative
f_prime = sp.diff(f, t)
print("Derivative Function:", f_prime)
# Turn the derivative into a Python function
f_prime_func = sp.lambdify(t, f_prime, 'numpy')

#Plot it
import matplotlib.pyplot as plt
import numpy as np
t_vals = np.linspace(-5, 5, 100)
plt.plot(t_vals, np.full_like(t_vals, f_func(t_vals)), color="black", label="f(t)",linewidth = 3)
plt.plot(t_vals, np.full_like(t_vals, f_prime_func(t_vals)), color="red",linestyle = '--', label="f'(t)")
plt.legend()
plt.show()
```
